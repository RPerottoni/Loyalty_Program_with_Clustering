```python
from sklearn import cluster as c

from sklearn.datasets import make_blobs
from yellowbrick.cluster import SilhouetteVisualizer, KElbowVisualizer
from matplotlib import pyplot as plt
```

```python
# Datasets

X, y = make_blobs( n_samples=500,
                   n_features=2,
                   centers=2,
                   cluster_std=0.5,
                   center_box=(-10, 10),
                   shuffle=True,
                   random_state=1)

# Plot Dataset
plt.scatter( X[:,0], X[:,1])
```

```python
# Modeling
clusters = [2,3,4,5,6,7,8]

kmeans=KElbowVisualizer(c.KMeans(),k=clusters, metric='silhouette', timings=False);
kmeans.fit( X );
kmeans.show();
```

```python
fix, ax = plt.subplots(2, 2, figsize=(12, 8))
for i in clusters:
    kmeans=c.KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100)
    kmeans.fit( X )

    q, mod = divmod(i, 2)
    ax[q-1, mod].scatter( X[:,0], X[:,1], c=kmeans.labels_, cmap='viridis')
    ax[q-1, mod].set_title('Number of Clusters: {}'.format( i ))
```

```python
fix, ax = plt.subplots(2, 2, figsize=(12, 12))
for i in clusters:
    kmeans=c.KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100)
    q, mod = divmod(i, 2)
   
    visualizer = SilhouetteVisualizer( kmeans, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit( X )
    visualizer.finalize()
```
